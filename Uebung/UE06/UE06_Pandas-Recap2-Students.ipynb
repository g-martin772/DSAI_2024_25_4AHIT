**Vorgehensweise**

Es wird ersichtlich, dass z.B. die Messung zum Zeitpunkt t2 fehlt. Hintergrund ist, dass all jene Messungen entfernt wurden, die mehr als **einen** NaN-Wert aufwiesen. Den zweiten und letzten Schritt, um diese Aufgabe zu bewerkstelligen, stellt die Ermittlung des Mittelwertes je Messzeitpunkt t dar. Das Ergebnis ist in die neue Spalte *mean* zu schreiben.

````python
path = "temperatures_with_NaN.csv"
df = pd.read_csv(path)
df.shape
````

````python
df[:5]
````

````python
df[["sensor3", "sensor4"]][:5]
````

````python
missing_values = df.isna().sum()
missing_values_formatted = missing_values.reset_index()
missing_values_formatted.columns = ['Column Name', 'Missing Count']
missing_values_formatted.sort_values(by='Missing Count', ascending=False, inplace=True)
missing_values_formatted
````

````python
df["mean"] = df.iloc[:, 1:].mean(axis=1)
df
````
