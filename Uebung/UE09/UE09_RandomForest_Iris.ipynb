```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
featureData = iris.data
targetLabels = iris.target
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(featureData, targetLabels, test_size=0.2, random_state=42)
```

```python
# Train a RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=300, max_depth=200, random_state=42)
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Visualize the feature importances
importances = classifier.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure()
plt.title("Feature importances")
plt.bar(range(featureData.shape[1]), importances[indices], align="center")
plt.xticks(range(featureData.shape[1]), iris.feature_names, rotation=90)
plt.xlim([-1, featureData.shape[1]])
plt.show()
```

```python
# Show confusion metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['setosa', 'versicolor', 'virginica'],
            yticklabels=['setosa', 'versicolor', 'virginica'])
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
# Map numbers to species names
species_names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}

for features, actual, predicted in zip(X_test, y_test, y_pred):
    print(f"{features} | {species_names[actual]} | {species_names[predicted]}")
```
